```python
import json
f = open('../../emerald/emerald_with_structured_abstr.jsonl', encoding='utf8')

abstract_tuples = {}
abtract_section_names = {}
section_names = {}
content_len = {}
path = {}

for line in f:
    j = json.loads(line.strip())
    abstract_tuples[tuple(j['abstract_sections_names'])] = abstract_tuples.get(tuple(j['abstract_sections_names']), 0) + 1
    for key in j['abstract_sections_names']:
        abtract_section_names[key] = abtract_section_names.get(key, 0) + 1
    for key in j['section_names']:
        section_names[key] = section_names.get(key, 0) + 1
    content_len[j['id']] = sum([i.count(' ') + 1 for i in j['section_names']]) +\
                            sum([j.count(' ') + 1 for i in j['sections'] for j in i])
    path[j['id']] = j['journal'] + ', ' + j['title']
```
